## Correlation Analysis
The purpose of this section is to analyze relationships between key variables in the K–12 school shootings dataset. I examined numerical correlations using Pearson correlation, Spearman correlation, and chi-square tests for categorical variables.

```python
import pandas as pd
import numpy as np
from scipy.stats import chi2_contingency

df = pd.read_csv('/Users/htallapr/Downloads/k12_school_shootings_merged_clean.csv')

numeric_df = df.select_dtypes(include=[np.number])
pearson_corr = numeric_df.corr(method='pearson')
pearson_corr
```

### Interpretation of Pearson Correlation
A correlation close to +1 indicates a strong positive relationship, -1 a strong negative relationship, and 0 no linear relationship.

```python
spearman_corr = numeric_df.corr(method='spearman')
spearman_corr
```

Pearson Correlation (Numeric Variables)

The Pearson correlation matrix showed very weak to no linear relationships between numeric variables such as:

Total_Victims
Victims_Killed
Victims_Wounded
Severity_Level
Duration_min

This indicates that severity levels (how many victims an incident involves) cannot be predicted from other numeric features in the dataset.
This makes sense because school shooting severity varies widely and depends on complex situational factors not captured numerically.

### Interpretation of Spearman Correlation
Spearman correlation measures ranked/monotonic relationships and is less sensitive to outliers.

```python
contingency_table = pd.crosstab(df['Motive_Category'], df['Location'])
chi2, p, dof, expected = chi2_contingency(contingency_table)
chi2, p
```

Spearman Correlation (Rank-Based Relationships)

Spearman correlation was also weak across all numeric variables, meaning that even if we only compare ranked values (higher vs. lower), there is still no consistent monotonic trend between severity and any other numeric variable.

Together, Pearson and Spearman both show that severity does not rise or fall in a predictable statistical pattern with other measurable features.

### Chi-Square Interpretation
- If p < 0.05: significant association.
- If p ≥ 0.05: no significant relationship.

```python
import matplotlib.pyplot as plt
import seaborn as sns

plt.figure(figsize=(22,17))
sns.heatmap(pearson_corr, annot=True, cmap='coolwarm')
plt.title('Pearson Correlation Heatmap')
plt.show()
```

Chi-Square Test (Motive vs. Location)

We used a chi-square test to analyze whether Motive_Category and Location are independent.

The chi-square test result showed a statistically significant association, meaning:

Certain motives tend to occur more often in specific types of school locations
(for example, disputes more common in hallways; targeted motives more common in classrooms).

This confirms that motive and location are not random — they have structured patterns.


Cramér’s V (Effect Size)
Although the chi-square test was significant, Cramér’s V was approximately 0.23, which indicates:
-A small-to-moderate relationship
-Motive and location are related, but not strongly
So while certain combinations are more common, the relationship is not strong enough to make hard predictions.

Kruskal–Wallis Test (Motive vs. Severity)
The Kruskal–Wallis test found no statistically significant difference between severity levels across different motive categories.
This means:
Incidents categorized under different motives
(targeted attack vs. accidental vs. gang-related)
do not show meaningful differences in the number of victims.

Severity is unpredictable and not tied to motive.
